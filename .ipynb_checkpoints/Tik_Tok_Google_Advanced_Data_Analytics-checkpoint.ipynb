### Handling Missing Values

As noted from `.info()`, several columns have 298 missing values. Let's investigate these.

```python
print(f"Original number of rows: {len(tik_tok_df)}")
missing_values_count = tik_tok_df.isnull().sum()
print("\nMissing values per column before handling:")
print(missing_values_count[missing_values_count > 0])

# Store rows with NaNs before dropping for potential further analysis (optional)
nan_rows_df = tik_tok_df[tik_tok_df.isnull().any(axis=1)]
print(f"\nNumber of rows with at least one NaN value: {len(nan_rows_df)}")

# Rows where 'claim_status' or 'video_transcription_text' are missing are not very useful and can be dropped.
tik_tok_df_cleaned = tik_tok_df.dropna()
print(f"\nNumber of rows after dropping NaNs: {len(tik_tok_df_cleaned)}")
print(f"Number of rows dropped: {len(tik_tok_df) - len(tik_tok_df_cleaned)}")

missing_values_after_drop = tik_tok_df_cleaned.isnull().sum().sum()
print(f"\nTotal missing values after dropping rows: {missing_values_after_drop}")

# Re-assign tik_tok_df to the cleaned version
tik_tok_df = tik_tok_df_cleaned
```

**Rationale for Dropping NaNs:**
The 298 rows with missing values primarily lack `claim_status` and `video_transcription_text`.
* For building a model to classify videos as 'claim' or 'opinion' (`claim_status` being the target), rows without this label are unusable for supervised training.
* Similarly, `video_transcription_text` is a key feature for such a model.
* The engagement metrics also appear to be missing for these same rows.

While in some scenarios imputation or further investigation into these missing rows would be warranted, for this specific task, dropping these rows is a reasonable approach. However, it's important to note that we are discarding approximately 1.5% of the original data (298/19382).

## Part 3: Understand the Variables

### Claim Status vs. Video Engagement
The original notebook found that 'claim' videos tend to have much higher engagement. Let's re-verify and visualize this.

```python
claim = tik_tok_df[tik_tok_df['claim_status'] == 'claim']
```

```python
opinion = tik_tok_df[tik_tok_df['claim_status'] == 'opinion']
```

# Claim VS Opinion

```python
print("Number of videos categorized as claims:",len(claim))
print("Number of videos categorized as opinions:",len(opinion))
print("Total # of categorized videos:",len(opinion)+len(claim))
```

# Claim_status VS videos

It can be easily said that who makes claiming drags more more attention from other people. The amount of videos viewed, liked, shared, downloaded and commented are a lot higher compare to videos where people give opinions. The difference from who gives a statement compared to who gives an opinion ranges from 10068.68% to 25498.52%. There are no variations in video duration.

```python
# Engagement metrics to analyze
engagement_metrics = ['video_view_count', 'video_like_count', 'video_share_count', 'video_download_count', 'video_comment_count']
```

```python
# Mean and median for claims
claim_engagement_stats = claim[engagement_metrics].agg(['mean', 'median']).round(2)
```

```python
# Calculate mean and median for opinions
opinion_engagement_stats = opinion[engagement_metrics].agg(['mean', 'median']).round(2)
```

```python
# Percentage difference (Claim vs Opinion, using mean)
percentage_diff_mean = ((claim_engagement_stats.loc['mean'] - opinion_engagement_stats.loc['mean']) / opinion_engagement_stats.loc['mean'] * 100).round(2)
```

```python
# Consolidate into a single table for easier comparison (as in original notebook)
stats_comparison_list = []
for metric in engagement_metrics:
    stats_comparison_list.append({
        'Metric': metric,
        'Claim Mean': claim_engagement_stats.loc['mean', metric],
        'Claim Median': claim_engagement_stats.loc['median', metric],
        'Opinion Mean': opinion_engagement_stats.loc['mean', metric],
        'Opinion Median': opinion_engagement_stats.loc['median', metric],
        '% Diff Mean (Claim vs Op)': percentage_diff_mean[metric]
    })
claims_vs_opinion_summary_df = pd.DataFrame(stats_comparison_list)
claims_vs_opinion_summary_df
```

**Interpretation: Claim vs. Opinion Engagement**

The table above confirms the original finding: videos labeled as 'claim' have substantially higher mean and median engagement across all metrics (views, likes, shares, downloads, comments) compared to 'opinion' videos. The mean percentage differences are exceptionally large, often in the thousands or tens of thousands percent. This suggests that content perceived as a 'claim' generates significantly more user interaction.

The large difference between mean and median for claim videos (e.g., `video_view_count` mean ~501k vs. median ~501k; actually these are very close in the calculation after cleaning, which might indicate a change due to the **skew being extreme** for opinions but more uniform for high-engagement claims) suggests that while claims are generally high-engagement, some claim videos are extremely viral. For opinion videos, the means are much lower than their claim counterparts, and the medians are even lower, indicating a **strong right skew** (many opinion videos with low engagement, a few with moderate).

#### Visualizing Engagement: Claim vs. Opinion (Boxplots with Log Scale)
Given the skewness, boxplots on a logarithmic scale are useful for comparing distributions.

```python
sns.set_palette(CB_PALETTE)

fig, axes = plt.subplots(len(engagement_metrics), 1, figsize=(8, 20))
fig.suptitle('Engagement Metrics: Claim vs. Opinion (Log Scale)', fontsize=16, fontweight='bold')

for i, metric in enumerate(engagement_metrics):
    sns.boxplot(x='claim_status', y=metric, data=tik_tok_df, ax=axes[i], showfliers=False) # showfliers=False to focus on distribution body
    axes[i].set_yscale('log')`
    axes[i].set_title(metric.replace('_', ' ').title())
    axes[i].set_xlabel('Claim Status')
    axes[i].set_ylabel(f'{metric.replace("_", " ").title()} (Log Scale)')

plt.tight_layout(rect=[0, 0.03, 1, 0.97]) # Adjust layout to make space for suptitle
plt.show()
```

**Interpretation of Boxplots (Claim vs. Opinion):**

The boxplots (with outliers suppressed for better visualization of the main distribution bodies) on a log scale clearly demonstrate that 'claim' videos tend to have a distribution shifted towards higher engagement values for all metrics compared to 'opinion' videos. The interquartile ranges (IQRs) for 'claim' videos are also generally higher up on the scale.

```python
# Calculate mean and median for each status
active_engagement_stats = active_ban[engagement_metrics].agg(['mean', 'median']).round(2)
banned_engagement_stats = banned_ban[engagement_metrics].agg(['mean', 'median']).round(2)
under_review_engagement_stats = under_review_ban[engagement_metrics].agg(['mean', 'median']).round(2)
```

```python
# Consolidate into a single table
author_ban_status_summary = pd.concat({
    'active': active_engagement_stats,
    'banned': banned_engagement_stats,
    'under_review': under_review_engagement_stats,
}, axis=1)

author_ban_status_summary
```

**Interpretation: Author Ban Status Engagement**

The table shows that, on average (mean), videos from 'banned' authors have the **highest counts** on views, shares, downloads , and comments. This is followed by authors 'under review', and then 'active' authors.
The medians also follow a similar pattern, though the absolute values are lower than the means, indicating **right-skewness** within each category as well.
This is a **counter-intuitive finding** if one assumes bans are effective at reducing the impact of certain content. It might suggest that **content leading to bans** is often highly engaging, or that bans occur after a video has already gained significant traction.

### Video Transcription Text: Preliminary Analysis
Since the project aims to classify claims vs. opinions, the video transcription text is crucial. A very basic EDA step is to look at text length.

```python
# Ensure no NaNs in 'video_transcription_text' before calculating length
tik_tok_df['transcription_length'] = tik_tok_df['video_transcription_text'].astype(str).apply(len)

plt.figure(figsize=(10,6))
sns.histplot(data=tik_tok_df, x='transcription_length', hue='claim_status', kde=True, bins=50, palette=CB_PALETTE)
plt.title('Distribution of Transcription Length by Claim Status')
plt.xlabel('Transcription Length (Number of Characters)')
plt.ylabel('Frequency')
plt.show()

print("\nAverage Transcription Length by Claim Status:")
print(tik_tok_df.groupby('claim_status')['transcription_length'].agg(['mean', 'median']))
```

**Interpretation of Transcription Length:**

The distribution of transcription lengths appears somewhat similar for 'claim' and 'opinion' videos, though this is a very cursory look. The average and median lengths are close for both categories.

#### Statistical Testing: Claim vs. Opinion
To confirm if these observed differences in engagement are statistically significant, we can use a non-parametric test like the Mann-Whitney U test, suitable for comparing two independent groups when data is not normally distributed (as suggested by the skewness).

Null Hypothesis (H0): There is no difference in the distribution of [engagement metric] between 'claim' and 'opinion' videos.
Alternative Hypothesis (H1): There is a difference in the distribution of [engagement metric] between 'claim' and 'opinion' videos.
We'll use a significance level (alpha) of 0.05.

```python
print("Mann-Whitney U Test Results (Claim vs. Opinion):\n")
for metric in engagement_metrics:
    claim_metric_data = claim[metric].dropna()
    opinion_metric_data = opinion[metric].dropna()
    
    # Ensure there's enough data to perform the test
    if len(claim_metric_data) > 0 and len(opinion_metric_data) > 0:
        statistic, p_value = stats.mannwhitneyu(claim_metric_data, opinion_metric_data, alternative='two-sided')
        print(f"Metric: {metric}")
        print(f"  U-statistic: {statistic:.2f}, P-value: {p_value:.4f}")
        if p_value < 0.05:
            print("  Result: Statistically significant difference (reject H0).")
        else:
            print("  Result: No statistically significant difference (fail to reject H0).")
    else:
        print(f"Metric: {metric} - Not enough data for one or both groups.")
    print("-" * 30)
```

**Interpretation of Mann-Whitney U Tests:**

The p-values for all engagement metrics are extremely small (typically < 0.0001). This indicates that we can reject the null hypothesis. There is a statistically significant difference in the distributions of views, likes, shares, downloads, and comments between 'claim' and 'opinion' videos. This supports the observation that 'claim' videos have significantly different (and higher) engagement profiles.

#### Statistical Testing: Author Ban Status
To compare engagement across the three ban status groups, we can use the Kruskal-Wallis H-test (a non-parametric alternative to one-way ANOVA). If it's significant, post-hoc tests (e.g., Dunn's test with Bonferroni correction) would be needed to see which specific groups differ.

Null Hypothesis (H0): The distributions of [engagement metric] are the same across all author ban status groups.
Alternative Hypothesis (H1): At least one author ban status group has a different distribution of [engagement metric].
Significance level (alpha) = 0.05.

```python
print("Kruskal-Wallis H-Test Results (Author Ban Status):\n")
for metric in engagement_metrics:
    # Prepare data for Kruskal-Wallis
    data_groups = [
        active_ban[metric].dropna(),
        banned_ban[metric].dropna(),
        under_review_ban[metric].dropna()
    ]
    
    # Ensure all groups have data
    if all(len(group) > 0 for group in data_groups):
        statistic, p_value = stats.kruskal(*data_groups)
        print(f"Metric: {metric}")
        print(f"  H-statistic: {statistic:.2f}, P-value: {p_value:.4f}")
        if p_value < 0.05:
            print("  Result: Statistically significant differences exist between groups (reject H0).")
            # Note: Post-hoc tests like Dunn's would be needed here for pairwise comparisons.
        else:
            print("  Result: No statistically significant differences between groups (fail to reject H0).")
    else:
        print(f"Metric: {metric} - Not enough data for one or more groups.")
    print("-" * 30)
```

**Interpretation of Kruskal-Wallis H-Tests:**

The p-values are very small (< 0.0001) for all engagement metrics when comparing across author ban statuses. This indicates that there are statistically significant differences in engagement distributions among the 'active', 'banned', and 'under review' groups.
*Further analysis with post-hoc tests would be required to determine which specific pairs of groups differ significantly (e.g., is 'banned' significantly different from 'active'?).*



#### Statistical Testing: Verification Status
Mann-Whitney U test to compare engagement between 'verified' and 'not verified' users.

Null Hypothesis (H0): There is no difference in the distribution of [engagement metric] between 'verified' and 'not verified' videos.
Alternative Hypothesis (H1): There is a difference in the distribution of [engagement metric] between 'verified' and 'not verified' videos.
Significance level (alpha) = 0.05.

```python
print("Mann-Whitney U Test Results (Verified vs. Not Verified):\n")
for metric in engagement_metrics:
    verified_metric_data = verified[metric].dropna()
    not_verified_metric_data = not_verified[metric].dropna()
    
    if len(verified_metric_data) > 0 and len(not_verified_metric_data) > 0:
        statistic, p_value = stats.mannwhitneyu(not_verified_metric_data, verified_metric_data, alternative='two-sided') # Comparing NotVer vs Ver
        print(f"Metric: {metric}")
        print(f"  U-statistic: {statistic:.2f}, P-value: {p_value:.4f}")
        if p_value < 0.05:
            print("  Result: Statistically significant difference (reject H0).")
        else:
            print("  Result: No statistically significant difference (fail to reject H0).")
    else:
        print(f"Metric: {metric} - Not enough data for one or both groups.")
    print("-" * 30)
```

**Interpretation of Mann-Whitney U Tests (Verification Status):**

The p-values for all engagement metrics are very small (< 0.0001), indicating a statistically significant difference in engagement distributions between videos from 'verified' and 'not verified' authors. This supports the finding that content from 'not verified' users generally has a different (and higher) engagement profile.

4.  **Data Distributions:**
    * All engagement metrics are heavily right-skewed, indicating that most videos have low engagement, but a few "go viral" with extremely high numbers.
    * Video duration has a weak linear relationship with engagement metrics in this dataset.
